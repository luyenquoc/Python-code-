```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.array([1,3,6])
y = np.array([4,6,8])
```

```python
a = np.arange(1,10)
b = a**4
```

```python
fig = plt.figure (figsize =(10,3),dpi = 200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[0][1].set_ylabel('Y Label 1,0')
axes[0][1].set_xlim(0,3)
axes[1][1].plot(a,b)
axes[1][1].set_title('Titel 1,1')
plt.tight_layout()
```
